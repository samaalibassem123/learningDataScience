# 3. update

### 3.1 Dropping Rows

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

### 3.2 Dropping Columns

```python
df.drop('Model Year', axis=1)
```

### 3.3 Creating Calculated Columns

```python
df['new Column'] = df['County']+ ',' + df['City']
```

```python
df.head()
```

### 3.4 Updating an Entire Column

```python
df['new Column'] = 100
```

```python
df.head()
```

### 3.5 Updating a Single Value

```python
df.iloc[0,-1] = 10
```

```python
df.head()
```

### 3.6 Condition Based Updating Using Apply

```python
df['City'] = df['Model Year'].apply(lambda x: 1 if x == 2020 else 0)
```
