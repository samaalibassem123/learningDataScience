# Learn Seaborn

```python
import pandas as pd
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
titanic = sns.load_dataset('titanic')
```

```python
df_titanic = pd.DataFrame(titanic)
df_titanic.index.name = 'index'
df_titanic.head()
```

```python
sns.scatterplot(x = 'index', y ='fare', data=df_titanic, hue='sex', size='alive' )
```

```python
sns.histplot(df_titanic['alive'], kde=True )
```

```python
sns.histplot(df_titanic['age'], bins=20)
```

```python
sns.displot(df_titanic['alive'])
```

```python
sns.barplot(x='sex', y='alive', data=df_titanic, hue = 'sex',palette='YlGnBu')
```

```python
sns.boxenplot(x='class', y = 'fare', data=df_titanic, hue='sex', palette='YlGnBu')
```

```python
sns.stripplot(x='class', y = 'fare', data=df_titanic, hue='sex', palette='YlGnBu')
```

```python
sns.jointplot(x = 'index', y = 'fare', data=df_titanic, kind='reg' )
```

```python
sns.jointplot(x = 'index', y = 'fare', data=df_titanic, kind='kde', fill=True, cmap='YlGnBu'  )
```

```python
sns.jointplot(x = 'index', y = 'fare', data=df_titanic, kind='hex', cmap='YlGnBu'  )
```

```python
sns.pairplot(df_titanic.select_dtypes(['number']), hue='pclass')
```
